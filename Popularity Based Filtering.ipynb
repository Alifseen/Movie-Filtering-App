## Calculate the Weighted Average Rating

A weighted average ensures that a movie that fewer people voted for is adjusted to be compared to movies with ratings by more users. So, a movie that got a 10/10 rating but has only 3 voters will be adjusted to be compared to a movie with a 7/10 rating from 3000 users.


WR = (v / (v+m)) * R + (m + (v + m)) * C  (formula from IMDB)

v = number of votes (this is the "vote_count" column)
m = minimum number of votes required to be considered (to be calculated)
R = average rating of the movie (this is the "vote_average" column)
C - average rating across all movies (to be calculated)

### Calculating m (Minimum Threshold)

```
m = movies['vote_count'].quantile(0.90)
```

### Calculating C (Avg Rating All movies)

```
C = movies['vote_average'].mean()
```

### Filter DataFrame based on minimum threshold

To avoid division by zero error, we create a filtered data frame before that only includes movies with more votes than the minimum threshold

```
filtered_movies = movies.copy().loc[movies['vote_count'] >= m]
```

Check new data frame

```
filtered_movies.head(10)
```

### Define a Function to calculate the weighted average

```
def weighted_rating(df, m=m, C=C):
    v = df['vote_count']
    R = df['vote_average']
    wr = (v/(v+m) * R) + (m/(m+v) * C)
    return wr
```

### Create a new weighted average column

```
filtered_movies['weighted_rating'] = filtered_movies.apply(weighted_rating, axis=1)
```

Sort the filtered data frame in descending order by the new values in a new column, the weighted rating

```
filtered_movies = filtered_movies.sort_values('weighted_rating', ascending=False)
```

Check Values

```
filtered_movies[['title', 'vote_count', 'vote_average', 'weighted_rating']].head(10)
```

### Convert DataFrame to Dictionary (Optional)

```
filtered_movies_dict = filtered_movies[['title', 'weighted_rating']].to_dict(orient='records')
```
